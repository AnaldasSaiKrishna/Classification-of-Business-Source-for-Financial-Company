```python
data=pd.get_dummies(data,columns=['Applicant_Gender','Applicant_Marital_Status','Applicant_Occupation','Applicant_Qualification','Manager_Current_Designation','Manager_Gender','Manager_Joining_Designation','Manager_Status'])
```

```python
trainm = data.loc[data['source']=="Train"]
testm  = data.loc[data['source']=="Test"]
```

```python
testm.drop(['Business_Sourced','source'],axis=1,inplace=True)
trainm.drop(['source','ID'],axis=1,inplace=True)
```

```python
trainm.to_csv("train_modified.csv",index=False)
testm.to_csv("test_modified.csv",index=False)
```
